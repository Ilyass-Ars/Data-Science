```python
def survie(pclass=2, sex=1, age=18):
    list_score=[]
    for i in range(1,11):
        model = KNeighborsClassifier(n_neighbors=i)
        model.fit(X,y)
        list_score.append(model.score(X,y))
        
    nmax = max(list_score)
    imax = list_score.index(nmax)
    model = KNeighborsClassifier(n_neighbors=imax)
    model.fit(X,y)
    
    x= np.array([pclass,sex,age]).reshape(1,3)
    plt.plot(list_score)
    print(f'max: {nmax}')
    print(f'{imax} = {model.predict(x)}')
    print(f'{imax} = {model.predict_proba(x)}')
```

## Model selection

```python
#train test split
#model.fit(X train, y train)
#model.score(X test, y test)
iris_set = load_iris()

X = iris_set.data
y = iris_set.target

print(X.shape)
plt.scatter(X[:,0],X[:,1],c=y, alpha=0.8)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
print('Train set', X_train.shape)
print('Test set',X_test.shape)
```

```python
model = KNeighborsClassifier(n_neighbors=6)
model.fit(X_train,y_train)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
# Train_set - Val_set( set de validation) - Test_set
#CrossValidation(entrain. et evalu. sur des proportion different)
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(KNeighborsClassifier( ),X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
#Validation Curve (idem cross validation avec test de plusieur hyperparam.)
from sklearn.model_selection import validation_curve
```

```python
model = KNeighborsClassifier()
k = np.arange(1,50)
train_score, val_score = validation_curve(model, X_train, y_train,param_name="n_neighbors", param_range=k,cv=5)

plt.plot(k,val_score.mean(axis = 1))

```

## GridSearchCV
- variation de plusieurs hyperparametres
- Grid= GridSearchCV(model,**param_grip**,cv)

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_neighbors': np.arange(1,20),
            'metric':['euclidean','manhattan']}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 5)
grid.fit(X_train,y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
print(grid.best_estimator_)
model = grid.best_estimator_
model.score(X_test,y_test)
```

## Metrique de regression
- MSE = (1/m)*Σ (y_vrai - y_predic)² quadratique
- MAE = (1/m)*Σ |y_vrai - y_predic| absolue
- RMSE = √(MSE) Root mean squared error
- MSE pénalise d'avantage les grande erreur que MAE
- MSE: si grande importance aux grandes erreurs.
- MAE: Si BD contient Outliers
- a mettre sur un graphe (plus comprehensible)

- Coefficient de determination R2, score(X,y) 0->1
- interpretation: le modele decrit 74% des variations du prix de ...

- CrossValidation: scoring='neg_mean_squared_error'
